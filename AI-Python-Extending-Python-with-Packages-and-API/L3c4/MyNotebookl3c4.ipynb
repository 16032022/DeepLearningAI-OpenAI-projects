One of the most popular Python third-party packages is called ```pandas```.

To use the ```pandas``` package, you first need to load it:

````python
import pandas as pd
````

RESPONSE CHATBOT : The line import pandas as pd in Python is an import statement that brings the Pandas library into the script and assigns it the alias pd.

Here's a breakdown:

import pandas: This imports the Pandas library, a powerful and widely used Python package for data manipulation and analysis.

as pd: This part creates an alias, meaning you can refer to the pandas library using the shorter name pd throughout your code. This is a common convention in the Python community to make the code more concise and readable.

Once Pandas is imported as pd, you can use it to perform tasks like:

Reading and writing data from various file formats (CSV, Excel, SQL databases, etc.)
Manipulating data using DataFrames (2D data structures) and Series (1D data structures)
Performing data cleaning, transformation, and analysis tasks
Working with time-series data
For example, after importing pandas as pd, you can load a CSV file into a DataFrame like this:

df = pd.read_csv('data.csv')


So `import pandas as pd` creates a shortcut that you can use to avoid typing pandas. 

**Note:** you'll need to use the `.` notation when using the function in pandas - go back to the "Using functions from a local file" lesson in this course if you need a refresher!

## Plotting data with ```matplotlib```

This is a popular package used by data scientists and developers to visualize data. You can check out the [Matplotlib documentation](https://matplotlib.org/) if you want to learn more!

To use matplotlib, you first have to import it with the following command:

````python
# Note the .pyplot
import matplotlib.pyplot as plt
````

Don't worry about remembering this command, you can always find it by asking a chatbot or searching online. Again, the `as` form of the import command is being used to set up an alias - so now you can type `plt` instead of `matplotlib.pyplot` saving you lots of typing!

Let's start by using ```matplotlib.pyplot``` to plot the selling price of a car against how many miles it has driven:

````python
plt.scatter(data["Kilometer"], data["Price"])

plt.title('Car Price vs. Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price (in USD)')

plt.show()
````

Ask the LLM to help you create a pie chart of the data showing how many cars were sold each year.

````python
import matplotlib.pyplot as plt

# Count the number of cars per year
year_counts = data['Year'].value_counts()

# Increase the size of the pie chart
plt.figure(figsize=(12,12)) #Adjust the figure size (width, height)

# Plot the pie chart
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Cars Sold per Year')
plt.show()
````
