#### Importing Functions from the statistics Package

````python
from statistics import mean, stdev
````

#### Calculating Mean and Standard Deviation
Create a list that contains the heights of a group of friends:

````python
# List of friends' heights
my_friends_heights = [160, 172, 155, 180, 165, 170, 158, 182, 175, 168]
````

You can now calculate the ```mean``` value of the list of heights using the `mean` function:

````python
# Calculate the mean value
mean_total = mean(my_friends_heights)
print(f"The mean value of height of the friends of Andrew is {mean_total}")
````

You can also calculate the standard deviation of the heights using the ```stdev``` function:

````python
# Calculate the standard deviation
stdev_total = stdev(my_friends_heights)
#stdev_total
print(f"The standard deviation of height of the friends of Andrew is {stdev_total}")
````

There are other functions in the statistics package that you can import to calculate other statistics, like median, etc. You can read more about the statistics package [here](https://docs.python.org/3/library/statistics.html) - or ask the chatbot!

#### Selecting Random Ingredients

Create lists of ingredients and use the sample function to select random items:

````python
# Lists of ingredients
spices = ["cumin", "turmeric", "oregano", "paprika"]
vegetables = ["lettuce", "tomato", "carrot", "broccoli"]
proteins = ["chicken", "tofu", "beef", "fish", "tempeh"]
````

You can use Python to select random ingredients for you from these lists. The ```sample``` function takes two parameters: the list you want to select from, and the number of items you want to select:

````python
# Select random items
random_spices = sample(spices, 2)  # Select 2 random items from the list
random_vegetables = sample(vegetables, 2) # Select 2 random items from the list
random_protein = sample(proteins, 1) # Select 1 random items from the list
````

#### Displaying Selected Ingredients

See what ingredients you selected:

````python
# Print selected ingredients
print(random_protein)
print(f"The random item that was selected from the given list is {random_protein}")
````

````python
# Print selected ingredients
print(random_spices)
print(f"The two random items that were selected from the given list are {random_spices}")
````

````python
# Print selected ingredients
print(random_vegetables)
print(f"The two random items that were selected from the given list are {random_vegetables}")
````
