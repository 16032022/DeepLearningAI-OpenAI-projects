### Plotting data with ```matplotlib```

This is a popular package used by data scientists and developers to visualize data. More info on  [Matplotlib documentation](https://matplotlib.org/).

To use `matplotlib`, you first have to import it with the following command:

````python
# Note the .pyplot
import matplotlib.pyplot as plt
````

**Note**  Again, the `as` form of the import command is being used to set up an alias - so now you can type `plt` instead of `matplotlib.pyplot` saving you lots of typing!

#### Creating a Scatter Plot
Let's start by using ```matplotlib.pyplot``` to plot the selling price of a car against how many miles it has driven:

````python
plt.scatter(data["Kilometer"], data["Price"])

plt.title('Car Price vs. Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price (in USD)')

plt.show()
````

#### Challenge exercise!

Ask the LLM to help you create a pie chart of the data showing how many cars were sold each year.

````python
# import matplotlib.pyplot as plt

# Count the number of cars per year
year_counts = data['Year'].value_counts()

# Create a threshold to group smaller segments into "Other"
threshold = 3  # Minimum percentage to keep separate
total = year_counts.sum()
filtered_counts = year_counts[year_counts / total * 100 >= threshold]
other_count = year_counts[year_counts / total * 100 < threshold].sum()

# Add the "Other" category if applicable
if other_count > 0:
    filtered_counts['Other'] = other_count

# Define a color palette and exploding effect
colors = plt.cm.tab20.colors  # Use a predefined colormap for clarity
explode = [0.1 if label == 'Other' else 0 for label in filtered_counts.index]  # Highlight "Other"

# Plot the improved pie chart
plt.figure(figsize=(10, 10))
plt.pie(filtered_counts, labels=filtered_counts.index, autopct='%1.1f%%', startangle=90,
        colors=colors[:len(filtered_counts)], explode=explode, textprops={'fontsize': 10})
plt.title('Percentage of Cars Sold per Year', fontsize=14)
plt.show()
````
